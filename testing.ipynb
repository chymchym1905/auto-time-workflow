```python
import re

def extract_resolution(formats):
    # Regular expression to match the number before 'p' and ensure 'p' follows the number
    pattern = re.compile(r'(\d+)p')

    resolutions = []
    for format in formats:
        match = pattern.search(format)
        if match:
            # Append the matched number (as an integer) to the resolutions list
            resolutions.append(int(match.group(1)))

    return resolutions

# Example list of video formats
formats = ["360p", "480p", "720p", "1080p60", "1440p60", "1440p30", "4k", "1080i"]

# Extract resolutions
resolutions = extract_resolution(formats)

print(resolutions)
```

```python
from datetime import datetime, timedelta
import calendar

def get_month_timestamps(year, month):
    # Start of the month
    start_of_month = datetime(year, month, 1, 0, 0, 0)
    
    # Determine the number of days in the month
    days_in_month = calendar.monthrange(year, month)[1]
    
    # End of the month (last second of the last day)
    end_of_month = datetime(year, month, days_in_month, 23, 59, 59)
    
    # Convert to Unix timestamps (in seconds)
    start_of_month_timestamp = int(start_of_month.timestamp())
    end_of_month_timestamp = int(end_of_month.timestamp())
    
    return start_of_month_timestamp, end_of_month_timestamp

def get_range_month_timestamps(start_year, start_month, end_year, end_month):
    current_year = start_year
    current_month = start_month
    timestamps = []
    
    while current_year < end_year or (current_year == end_year and current_month <= end_month):
        start_ts, end_ts = get_month_timestamps(current_year, current_month)
        timestamps.append((start_ts, end_ts))
        
        # Move to the next month
        if current_month == 12:
            current_month = 1
            current_year += 1
        else:
            current_month += 1
    
    return timestamps

# Example usage
start_year = 2023
start_month = 3
end_year = 2023
end_month = 3

timestamps = get_range_month_timestamps(start_year, start_month, end_year, end_month)

for start_ts, end_ts in timestamps:
    print(f"Start: {datetime.fromtimestamp(start_ts).strftime('%Y-%m-%d %H:%M:%S')}, Unix timestamp: {start_ts}")
    print(f"End: {datetime.fromtimestamp(end_ts).strftime('%Y-%m-%d %H:%M:%S')}, Unix timestamp: {end_ts}")
    print()
```

```python
from datetime import datetime, timedelta

def get_day_timestamps(year, month, day):
    # Start of the day
    start_of_day = datetime(year, month, day, 0, 0, 0)
    # End of the day
    end_of_day = datetime(year, month, day, 23, 59, 59)
    
    # Convert to Unix timestamps (in seconds)
    start_of_day_timestamp = int(start_of_day.timestamp())
    end_of_day_timestamp = int(end_of_day.timestamp())
    
    return start_of_day_timestamp, end_of_day_timestamp

def get_range_day_timestamps(start_year, start_month, start_day, end_year, end_month, end_day):
    start_date = datetime(start_year, start_month, start_day)
    end_date = datetime(end_year, end_month, end_day)
    current_date = start_date
    timestamps = []
    
    while current_date <= end_date:
        start_ts, end_ts = get_day_timestamps(current_date.year, current_date.month, current_date.day)
        timestamps.append((start_ts, end_ts))
        
        # Move to the next day
        current_date += timedelta(days=1)
    
    return timestamps

# Example usage
start_year = 2023
start_month = 4
start_day = 4
end_year = 2023
end_month = 5
end_day = 4

timestamps = get_range_day_timestamps(start_year, start_month, start_day, end_year, end_month, end_day)

for start_ts, end_ts in timestamps:
    print(f"Start: {datetime.fromtimestamp(start_ts).strftime('%Y-%m-%d %H:%M:%S')}, Unix timestamp: {start_ts}")
    print(f"End: {datetime.fromtimestamp(end_ts).strftime('%Y-%m-%d %H:%M:%S')}, Unix timestamp: {end_ts}")
    print()
```
